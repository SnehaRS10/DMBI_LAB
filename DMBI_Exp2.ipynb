```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Global YouTube Statistics.csv', encoding='iso-8859-1')
data
```

```python
data.isna().sum()
```

```python
category = []
for cat in data:
    if data[cat].dtype == 'O':
        category.append(cat)
        -
category
```

```python
categorical = data[category].dropna()
categorical
```

```python
youtube= pd.merge(data, categorical)
youtube
```

```python
youtube.isna().sum()
```

```python
cols_to_drop= ['subscribers_for_last_30_days', 'video_views_for_the_last_30_days']
```

```python
youtube.drop(cols_to_drop, axis=1, inplace=True)
youtube.isna().sum()
```

```python
you1 = youtube[['Youtuber','subscribers' ,'video views', 'category', 'uploads', 'Country', 'channel_type', 'Population', 'Unemployment rate', 'highest_monthly_earnings']]
you1
```

```python
you1.describe()
```

```python
# There is a strong correlation between video views and subscribers
you1['video views'].corr(you1['subscribers'])
```

```python
you1.plot.scatter(x='video views', y='subscribers')
plt.xlabel('Video Views')
plt.ylabel('Subscribers')
plt.title('Relationship Between Subscribers and Video Views')
```

```python
# There is a weak correlation between video views and monthly earnings.
# This because other factors affect the monthly earnings
you1['highest_monthly_earnings'].corr(you1['video views'])
```

```python
you1.plot.scatter(x='video views', y='highest_monthly_earnings')
plt.xlabel('Video Views')
plt.ylabel('Highest Monthly Earnings')
plt.title('Relationship Between Video Views and Highest Monthly Earnings')
```

```python
you1['highest_monthly_earnings'].corr(you1['subscribers'])
```

```python
you1.plot.scatter(x='subscribers', y='highest_monthly_earnings')
plt.xlabel('Subscribers')
plt.ylabel('Highest Monthly Earnings')
plt.title('Relationship Between Subscriberss and Highest Monthly Earnings')
```

```python
Earnings_by_category = you1.groupby(['category'])['highest_monthly_earnings'].sum()
Earnings_by_category
```

```python
Earnings_by_category.plot.bar()
plt.xlabel('category')
plt.ylabel('highest_monthly_earnings')
plt.title('Monthly Earnings by Category')
```

```python
you1.plot.scatter(x='Population', y='subscribers')
plt.xlabel('Population')
plt.ylabel('Subscribers')
plt.title('Relationship Between Video views and Subscribers')
```

```python
you1['Population'].corr(you1['subscribers'])
```

```python
you1.plot.scatter(x='Population', y='video views')
plt.xlabel('Population')
plt.ylabel('Video views')
plt.title('Relationship Between Popuation and Video views')
```

```python
you1['Unemployment rate'].corr(you1['subscribers'])
```
